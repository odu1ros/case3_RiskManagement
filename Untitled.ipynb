```python
data = pd.read_excel('data/insurance_claims.xlsx')
data.head()
```

```python
data.columns
```

```python
data = data.drop(['incident_date', 'policy_date', 'job_start_date', 'occupancy_date', 'dob'], axis=1)
data.head()
```

```python
for i in range(len(data)):
    if data.iloc[i, list(data.columns).index('worked_til_incident')] < 0:
        data.iloc[i, list(data.columns).index('worked_til_incident')] = 0
        
for i in range(len(data)):
    if data.iloc[i, list(data.columns).index('lived_til_incident')] < 0:
        data.iloc[i, list(data.columns).index('lived_til_incident')] = 0
```

```python
data.describe()
```

```python
data['edcat'] = data['edcat'].replace([2, 3, 5], 0)
data['edcat'] = data['edcat'].replace([1, 4], 1)

data['claim_type'] = data['claim_type'].replace([1, 2], 0)
data['claim_type'] = data['claim_type'].replace([3, 5], 1)
data['claim_type'] = data['claim_type'].replace([4], 2)

data['townsize'] = data['townsize'].replace([1, 2], 0)
data['townsize'] = data['townsize'].replace([3, 5], 1)
data['townsize'] = data['townsize'].replace([4], 2)

data['reside'] = data['reside'].replace([2, 5, 6], 0)
data['reside'] = data['reside'].replace([1, 3, 4], 1)
data['reside'] = data['reside'].replace([7, 8, 9, 10], 2)

for i in range(len(data)):
    if data.iloc[i, data.columns.get_loc('deductible')] >= 250:
        data.iloc[i, data.columns.get_loc('deductible')] = 1
    else:
        data.iloc[i, data.columns.get_loc('deductible')] = 0
```

```python
data = data.drop(['deductible', 'uninhabitable', 'marital', 'edcat', 'gender', 'primary_residence', 'retire'], axis=1)
```

```python
y = data_analysis.iloc[:, data_analysis.columns == 'fraudulent']
X = data_analysis.iloc[:, data_analysis.columns != 'fraudulent']

y1 = data_prediction.iloc[:, data_prediction.columns == 'fraudulent']
X1 = data_prediction.iloc[:, data_prediction.columns != 'fraudulent']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
model = xgb.XGBRegressor()
model.fit(X_train, y_train)

yhat = model.predict(X_test)

print('R-squared score (training): {:.3f}'.format(model.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'.format(model.score(X_test, y_test)))
print('ROC AUC score (test): {:.3f}'.format(roc_auc_score(y_test, yhat, multi_class='ovr')))
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=y)

xgb.plot_importance(model)
plt.rcParams['figure.figsize'] = [5, 5]
plt.show()
```
